```
# Model 1:
# Things Tried: 1. Transition Layer 2. Maxpooling
# Accuracy: 98.64 (10 epochs)

class Net(nn.Module):
    def __init__(self):
      super(Net, self).__init__()
      
      self.conv1 = nn.Conv2d(1, 8, 3) #input-28 | Output-26 | RF-3

      self.conv2 = nn.Conv2d(8, 8, 3) #input-26 | Output-24 | RF-5
      self.conv3 = nn.Conv2d(8, 16, 3) #input-24 | Output-22 | RF-7

      self.pool1 = nn.MaxPool2d(2, 2) #input-22 | Output-11 | RF-14
      self.conv4 = nn.Conv2d(16, 8, 1) #input-11 | Output-11 | RF-14
      
      self.conv5 = nn.Conv2d(8, 8, 3) #input-11 | Output-9 | RF-16
      self.conv6 = nn.Conv2d(8, 16, 3) #input-9 | Output-7 | RF-18
      self.conv7 = nn.Conv2d(16, 32, 3) #input-7 | Output-5 | RF-20

      self.conv8 = nn.Conv2d(32, 10, 5) #input-7 | Output-5 | RF-20


    def forward(self, x):
      x = F.relu(self.conv1(x))
      x = F.relu(self.conv2(x))
      x = F.relu(self.conv3(x))

      x = self.pool1(x)

      x = F.relu(self.conv4(x))
      x = F.relu(self.conv5(x))
      x = F.relu(self.conv6(x))
      x = F.relu(self.conv7(x))

      x = F.relu(self.conv8(x))

      x = x.view(-1, 10)
      return F.log_softmax(x)
```

```
# Model 4:
# Things Tried: 1. Transition Layer 2. Maxpooling
# Accuracy: 98.8% after 20 epochs

class Net(nn.Module):
    def __init__(self):
      super(Net, self).__init__()
      
      self.conv1 = nn.Conv2d(1, 8, 3)     #input-28 | Output-26 | RF-3
      self.conv2 = nn.Conv2d(8, 8, 3)     #input-26 | Output-24 | RF-5
      self.conv3 = nn.Conv2d(8, 8, 3)     #input-24 | Output-22 | RF-7

      self.pool1 = nn.MaxPool2d(2, 2)     #input-22 | Output-11 | RF-14
      
      self.conv4 = nn.Conv2d(8, 16, 3)    #input-11 | Output-09 | RF-16
      self.conv5 = nn.Conv2d(16, 16, 3)   #input-09 | Output-07 | RF-18
      self.conv6 = nn.Conv2d(16, 16, 3)   #input-07 | Output-05 | RF-20

      self.conv7 = nn.Conv2d(16, 10, 5)   #input-04 | Output-01 |


    def forward(self, x):
      x = F.relu(self.conv1(x))
      x = F.relu(self.conv2(x))
      x = F.relu(self.conv3(x))

      x = self.pool1(x)

      x = F.relu(self.conv4(x))
      x = F.relu(self.conv5(x))
      x = F.relu(self.conv6(x))
      x = F.relu(self.conv7(x))
      
      x = x.view(-1, 10)
      
      return F.log_softmax(x)
```

```
# Model 5:
# Things Tried: 1. Transition Layer 2. Maxpooling 3. Dropout
# Accuracy: 99.2% after 20 epochs
# Params = 11,066

class Net(nn.Module):
    def __init__(self):
      super(Net, self).__init__()
      
      self.conv1 = nn.Conv2d(1, 8, 3)     #input-28 | Output-26 | RF-3
      self.do1 = nn.Dropout(0.1)

      self.conv2 = nn.Conv2d(8, 8, 3)     #input-26 | Output-24 | RF-5
      self.do2 = nn.Dropout(0.1)

      self.conv3 = nn.Conv2d(8, 8, 3)     #input-24 | Output-22 | RF-7
      self.do3 = nn.Dropout(0.1)

      self.pool1 = nn.MaxPool2d(2, 2)     #input-22 | Output-11 | RF-14
      
      self.conv4 = nn.Conv2d(8, 16, 3)    #input-11 | Output-09 | RF-16
      self.do4 = nn.Dropout(0.1)

      self.conv5 = nn.Conv2d(16, 16, 3)   #input-09 | Output-07 | RF-18
      self.do5 = nn.Dropout(0.1)

      self.conv6 = nn.Conv2d(16, 16, 3)   #input-07 | Output-05 | RF-20
      self.do6 = nn.Dropout(0.1)

      self.conv7 = nn.Conv2d(16, 10, 5)   #input-04 | Output-01 |


    def forward(self, x):
      x = F.relu(self.conv1(x))
      x = self.do1(x)

      x = F.relu(self.conv2(x))
      x = self.do2(x)

      x = F.relu(self.conv3(x))
      x = self.do3(x)

      x = self.pool1(x)

      x = F.relu(self.conv4(x))
      x = self.do4(x)

      x = F.relu(self.conv5(x))
      x = self.do5(x)

      x = F.relu(self.conv6(x))
      x = self.do6(x)

      x = F.relu(self.conv7(x))

      x = x.view(-1, 10)
      
      return F.log_softmax(x)
```

```
# Model 6:
# Things Tried: 1. Transition Layer 2. Maxpooling 3. Dropout 4. Batch Normalization
# Accuracy: 99.4% in 18th epoch
# Params = 11,066

class Net(nn.Module):
    def __init__(self):
      super(Net, self).__init__()
      
      self.conv1 = nn.Conv2d(1, 8, 3)     #input-28 | Output-26 | RF-3
      self.bn1 = nn.BatchNorm2d(8)
      self.do1 = nn.Dropout(0.1)

      self.conv2 = nn.Conv2d(8, 8, 3)     #input-26 | Output-24 | RF-5
      self.bn2 = nn.BatchNorm2d(8)
      self.do2 = nn.Dropout(0.1)

      self.conv3 = nn.Conv2d(8, 8, 3)     #input-24 | Output-22 | RF-7
      self.bn3 = nn.BatchNorm2d(8)
      self.do3 = nn.Dropout(0.1)

      self.pool1 = nn.MaxPool2d(2, 2)     #input-22 | Output-11 | RF-14
      
      self.conv4 = nn.Conv2d(8, 16, 3)    #input-11 | Output-09 | RF-16
      self.bn4 = nn.BatchNorm2d(16)
      self.do4 = nn.Dropout(0.1)

      self.conv5 = nn.Conv2d(16, 16, 3)   #input-09 | Output-07 | RF-18
      self.bn5 = nn.BatchNorm2d(16)
      self.do5 = nn.Dropout(0.1)

      self.conv6 = nn.Conv2d(16, 16, 3)   #input-07 | Output-05 | RF-20
      self.bn6 = nn.BatchNorm2d(16)
      self.do6 = nn.Dropout(0.1)

      self.conv7 = nn.Conv2d(16, 10, 5)   #input-04 | Output-01 |


    def forward(self, x):
      x = F.relu(self.conv1(x))
      x = self.bn1(x)
      x = self.do1(x)

      x = F.relu(self.conv2(x))
      x = self.bn2(x)
      x = self.do2(x)

      x = F.relu(self.conv3(x))
      x = self.bn3(x)
      x = self.do3(x)

      x = self.pool1(x)

      x = F.relu(self.conv4(x))
      x = self.bn4(x)
      x = self.do4(x)

      x = F.relu(self.conv5(x))
      x = self.bn5(x)
      x = self.do5(x)

      x = F.relu(self.conv6(x))
      x = self.bn6(x)
      x = self.do6(x)

      x = F.relu(self.conv7(x))

      x = x.view(-1, 10)
      
      return F.log_softmax(x)
```
